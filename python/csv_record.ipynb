```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_file(filename):
    # Load CSV
    df = pd.read_csv(filename)

    time = df["time"]
    voltage = df["voltage"]
    current = df["current"]
    power = df["total_power"]

    # Create 4 plots in one row
    fig, axs = plt.subplots(1, 4, figsize=(20, 4))

    fig.suptitle(filename, fontsize=16)

    # Voltage plot
    axs[0].plot(time, voltage, color="green")
    axs[0].set_title("Voltage")
    axs[0].set_xlabel("Time (s)")
    axs[0].set_ylabel("Voltage (V)")
    axs[0].set_ylim(0, 12.5)

    # Current plot
    axs[1].plot(time, current, color="blue")
    axs[1].set_title("Current")
    axs[1].set_xlabel("Time (s)")
    axs[1].set_ylabel("Current (A)")
    axs[1].set_ylim(0, 6)

    # Power plot
    axs[2].plot(time, power, color="red")
    axs[2].set_title("Power")
    axs[2].set_xlabel("Time (s)")
    axs[2].set_ylabel("Power (W)")
    axs[2].set_ylim(0, 150)

    # Combined plot
    axs[3].plot(time, voltage, color="green", label="Voltage")
    axs[3].plot(time, current, color="blue", label="Current")
    axs[3].plot(time, power, color="red", label="Power")
    axs[3].set_title("All Signals")
    axs[3].set_xlabel("Time (s)")
    axs[3].set_ylim(0, 150)
    axs[3].legend()

    plt.tight_layout()
    plt.show()


# Plot both files
plot_file("bootup.csv")
plot_file("benchmark.csv")
```

```python
import numpy as np
df = pd.read_csv("benchmark.csv")

time = df["time"]
voltage = df["voltage"]
current = df["current"]
power = df["total_power"]

# Create 4 plots in one row
fig, axs = plt.subplots(1, 1, figsize=(19, 6))

fig.suptitle("benchmark.csv", fontsize=16)

# Power plot
axs.plot(time, power, color="red")
axs.set_title("Power")
axs.set_xlabel("Time (s)")
axs.set_ylabel("Power (W)")
axs.set_ylim(0, 150)
axs.axvline(x=40, color='black', linestyle='--', linewidth=1)
axs.axvline(x=54, color='black', linestyle='--', linewidth=1)
axs.axvline(x=70, color='black', linestyle='--', linewidth=1)
axs.axvline(x=86, color='black', linestyle='--', linewidth=1)
axs.set_xticks(np.linspace(0, 220, 23))

plt.tight_layout()
plt.show()
```
